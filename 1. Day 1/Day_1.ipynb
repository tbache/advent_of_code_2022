# Advent of Code 2022 - Day 1

```python
import numpy as np
```

```python
with open('input.txt', 'r') as f:
    data = f.read().split('\n\n')
```

```python
data[0:100]
```

```python
data[0:10]
```

```python
# Add up each elf's calories
totalCals = []
for i in data:
    totalCals.append(sum([int(x) for x in i.split()]))
```

```python
totalCals = np.array(totalCals)
```

## Part 1

```python
# Find the elf with the most calories
elf_with_most = np.max(totalCals)
```

```python
elf_with_most
```

## Part 2

```python
# Find the total calories for the three elves with the most calories
sum(sorted(totalCals)[-3:])
```
